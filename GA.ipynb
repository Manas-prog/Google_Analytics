```python
import json
import os
import pandas as pd
```

```python

# Folder path where your JSON files are stored
folder_path = 'D:\\MSBA\\GA\\'

# List to store all records from all files
all_data = []

# Loop through each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, 'r', encoding='utf-8') as file:
            for line in file:
                try:
                    all_data.append(json.loads(line))
                except json.JSONDecodeError as e:
                    print(f"Skipping invalid line in {filename}:", e)

# Convert list of records to DataFrame
df = pd.DataFrame(all_data)

# Display results
print("Total records loaded:", len(df))
```

```python
# Extract multiple parameters at once in a more optimized way
def extract_params(event_params, param_keys):
    """Extract multiple parameters from event_params"""
    results = {key: None for key in param_keys}
    
    for param in event_params:
        key = param.get('key')
        if key in param_keys:
            value_dict = param.get('value', {})
            
            # Extract based on value type
            if 'string_value' in value_dict:
                results[key] = value_dict['string_value']
            elif 'int_value' in value_dict:
                results[key] = value_dict['int_value']
            elif 'float_value' in value_dict:
                results[key] = value_dict['float_value']
            elif 'double_value' in value_dict:
                results[key] = value_dict['double_value']
    
    return results

# Step 5: Define parameters we want to extract
params_to_extract = ['page_title', 'page_location', 'source', 'medium', 'ga_session_id', 'engagement_time_msec', 'session_engaged']

# Step 6: Apply the extraction function to each row
extracted_data = df_1['event_params'].apply(lambda x: extract_params(x, params_to_extract))

# Step 7: Convert the result to a DataFrame
params_df = pd.DataFrame(extracted_data.tolist())
print("Extracted parameters:")
print(params_df.head(3))
print("\n")
```
